```python
def setupjointpdf( vpdata, svpdata, weights ):
    #X, Y = np.mgrid[ -2:4:241j, -2:4:241j ] # probabilities represent per (0.025hPa x 0.025hPa grid box)
    #X, Y = np.mgrid[ -2:5:281j, -2:5:281j ] # probabilities represent per (0.025hPa x 0.025hPa grid box)
    #X, Y = np.mgrid[-1.5:1.525:120j, -2.5:5:75j] # probabilities represent per (0.025 precip, 0.1 vp)
    X, Y = np.mgrid[-10:10.1:200j, -10:10.1:200j]
    positions = np.vstack([X.ravel(), Y.ravel()])
    
    xplot = vpdata.flatten()
    yplot = svpdata.flatten()
    inputdata = np.vstack([xplot, yplot])
    kernel = stats.gaussian_kde(inputdata, weights=weights)
    Z = np.reshape(kernel(positions).T, X.shape)
    Z = np.moveaxis(Z, 1, 0)
    Xout = X[:,0] ; Yout = Y[0,:]
    return Z, Xout, Yout
```

```python
w = np.cos(np.deg2rad(datarid.lats))
zarid, x, y = setupjointpdf(np.array(datarid.isd), np.array(datarid.era5), w)
w = np.cos(np.deg2rad(dathumid.lats))
zhumid, x, y = setupjointpdf(np.array(dathumid.isd), np.array(dathumid.era5), w)
w = np.cos(np.deg2rad(dathyperarid.lats))
zhyperarid, x, y = setupjointpdf(np.array(dathyperarid.isd), np.array(dathyperarid.era5), w)


w = np.cos(np.deg2rad(datarid_hadisdh.lat))
zarid_hadisdh, x, y = setupjointpdf(np.array(datarid_hadisdh.hadisdh), np.array(datarid_hadisdh.era5), w)
w = np.cos(np.deg2rad(dathumid_hadisdh.lat))
zhumid_hadisdh, x, y = setupjointpdf(np.array(dathumid_hadisdh.hadisdh), np.array(dathumid_hadisdh.era5), w)
w = np.cos(np.deg2rad(dathyperarid_hadisdh.lat))
zhyperarid_hadisdh, x, y = setupjointpdf(np.array(dathyperarid_hadisdh.hadisdh), np.array(dathyperarid_hadisdh.era5), w)
```
